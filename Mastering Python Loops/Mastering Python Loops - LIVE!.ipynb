**Given a list, print all the even numbers in the list**

```python
list1 = [2, 3, 6, 8, 12, 13, 17]
for num in list1:
  if num % 2 == 0:
    print(num)
```

```python
list1 = [2, 3, 6, 8, 12, 13, 17]
list2 = []
for num in list1:
  if num % 2 == 0:
    list2.append(num)

print(list2)
```

**Summation of all the values available in the given list**

```python
sum = 0
for num in list1:
  sum = sum + num

print(sum)
```

**Implementation of the same code via applying the for loop**

```python
num = 6
factorial = 1
while num > 0:
  factorial = factorial * num
  num -= 1
print(factorial)
```

**Two Pointer Technique**

```python
list2 = [2, 5, 8, 12, 17]
low = 0
high = len(list2) - 1
while low < high:
  ## swap list2[low] with the list2[high]
  list2[low], list2[high] = list2[high], list2[low]
  ## increment the low pointer and decrement the high pointer
  low += 1
  high -= 1

print(list2)
```
