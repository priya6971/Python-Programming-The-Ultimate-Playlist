```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = sns.load_dataset('tips')
```

```python
data.head()
```

```python
data.shape
```

**Visualize the relationship between tip and the total_bill value.**

```python
sns.relplot(x='tip', y='total_bill', data=data)
```

```python
sns.relplot(x='tip', y='total_bill', data=data, hue='day')
```

```python
sns.catplot(x='day', y='tip', data=data)
```

```python
sns.catplot(x='smoker', y='tip', data=data)
```

```python
sns.catplot(x='day', y='tip', data=data, kind='box', hue='sex')
```

```python
sns.distplot(data.tip)
```
