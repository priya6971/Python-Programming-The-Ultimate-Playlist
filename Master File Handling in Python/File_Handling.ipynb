Open the file

"r" -> read mode which helps to read the file in one go

read() -> reads the entire content of the file and returns it as a single string.

```python
file = open("/content/demo_file.txt", "r")
# read the entire file
content = file.read()
print(content)
```

read only a specific part of the content available in the demo file.

```python
file = open("/content/demo_file.txt", "r")
# return the first 6 characters of the file
print(file.read(6))
```

readline() -> read one line from the file at a time.

```python
file = open("/content/demo_file.txt", "r")
# read the first line
line1 = file.readline()
print("First line:", line1)
```

```python
# read the second line
line2 = file.readline()
print("Second line:", line2)
```

```python
# read the third line
line3 = file.readline()
print("Third line:", line3)
```

readlines() -> returns all the lines of the file into a list of strings.

```python
file = open("/content/demo_file.txt", "r")
# read all the lines and store them into a list
line = file.readlines()
print("All lines: ", line)
```

Closing the file

```python
file = open("/content/demo_file.txt", "r")
# file operations ...
file.close() # close the file explicitly
```

finally block: always run without considering whether your try block is running or your except block is running.

```python
file = open("/content/demo_file.txt", "r")
try:
  # file operations ...
  pass
finally:
  file.close()
```

```python
file = open("/content/demo_file.txt", "r")
try:
  content = file.read()
  print(content)
except IOError:
  print("Error: An IOError occurred")
except Exception:
  print("An unexpected error occurred")
finally:
  file.close()
```

"w" -> write mode which is specifically useful for writing in the file. If the file doesn't exist, it creates a new file.

If the file already exists, it overwrites the content.

```python
file = open("/content/demo_file.txt", "w")
content = file.write("This is a sample text used to demonstrate the concepts of file handling.\n")
print(content)
file.close()
```

```python
file = open("/content/demo_file.txt", "r")
# read the entire file
content = file.read()
print(content)
```

"a" -> append mode opens a file and append the data at the end of the file. If the file doesn't exist, it creates a new file.

```python
file = open("/content/demo_file.txt", "a")
file.write("I hope you will enjoy the today's demonstration.\n")
file.write("And find the overall discussion very helpful and insightful for your upcoming career.\n")
file.write("Happy Learning!")
```

```python
file = open("/content/demo_file.txt", "r")
# read the entire file
content = file.read()
print(content)
```

"x" -> Exclusive creation mode

This means that the file will be created only if it does not already exists. If the file already exists, then the operation fails and raises a FileExistsError.

```python
file = open("/content/new_file.txt", "x")
```

```python
file.write('This is a new file specially created using x mode.\n')
file.write('Writing to this file will fail if the file already exists.')
```

```python
file = open("/content/new_file.txt", "r")
# read the entire file
content = file.read()
print(content)
```
