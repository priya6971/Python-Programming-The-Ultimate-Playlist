```python
class TechforAllWithPriya:
  ## initialization
  def __init__(self, courseId, courseName, instructorName):
    self.courseID = courseId
    self.course_Name = courseName
    self.instructor = instructorName

  ## to get the details of course
  def course_details(self):
    return f"{self.courseID} {self.course_Name}"

  ## to display the complete data
  ## concept of polymorphism
  def display_details(self):
    details = "Course ID: {}\nCourse Name: {}\n Instructor Name: {}".format(self.courseID, self.course_Name, self.instructor)
    return details

## inheritance
## Learners: Derived class
## TechforAllWithPriya: Parent class
class Learners(TechforAllWithPriya):
  ## initialization
  def __init__(self, courseId, courseName, instructorName, studentId, studentName):
    super().__init__(courseId, courseName, instructorName)
    self.studentId = studentId
    self.student_Name = studentName

  ## to display the complete data
  ## concept of polymorphism
  def display_details(self):
    parent_details = super().display_details()
    details = "{}\n Student ID: {}\n Student Name: {}".format(parent_details, self.studentId, self.student_Name)
    return details

## object of TechforAllWithPriya class
sqlcourse = TechforAllWithPriya(1, "7-Days SQL Bootcamp", "Priya Bhatia")
print(sqlcourse.course_Name)
print(sqlcourse.instructor)
print(sqlcourse.courseID)
print(sqlcourse.course_details())
print(sqlcourse.display_details())

## object of Learners class
enrolledLearner = Learners(1, "7-Days SQL Bootcamp", "Priya Bhatia", 17, "Akhil")
print(enrolledLearner.student_Name)
print(enrolledLearner.studentId)
print(enrolledLearner.display_details())
```
